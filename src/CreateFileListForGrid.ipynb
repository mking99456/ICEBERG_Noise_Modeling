```python
#This notebook will search through the run CSV and create a list of the files
#Which all have the same characteristics:
#
#Pulser mode true
#Pulser DAC 0
#Gain 2
#Shaping time 3
```

```python
import csv
goodfiles = []
totalevents = 0
with open( '/home/matt/Documents/Grad School/Research/ICEBERG_Project/Run_Spreadsheets/Run5_Iceberg_config_spreadsheet_apr5_2021.csv', 'r' ) as theFile:
    reader = csv.DictReader(theFile)
    for line in reader:
        # line is { 'workers': 'w0', 'constant': 7.334, 'age': -1.406, ... }
        # e.g. print( line[ 'workers' ] ) yields 'w0'
        print(line)
        if (line[ ' Pulser Mode' ] == ' true' and line[ ' Pulser DAC' ] == '0' and line[ ' Gain' ] =='2' and line[' Shaping Time']=='3' and line[' Baseline High'] == '0'):
            goodfiles.append(line[' Data file name'])
        if (int(line[' Run Number']) >= 9693 and int(line[' Run Number']) <= 9699) or (int(line[' Run Number']) >= 9701 and int(line[' Run Number']) <= 9711):
            totalevents += int(line[' Event Count'])
print(goodfiles)
print(totalevents)


```

```python
#Now create the text file
print(goodfiles)
print(len(goodfiles))

with open('runfiles.txt','w') as f:
    for line in goodfiles:
        f.write(line+'\n')
```
